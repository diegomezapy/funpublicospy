## Explorar `compiled_df`

### Subtask:
Explorar la estructura y el contenido del DataFrame `compiled_df` mostrando las primeras filas, la informaciÃ³n de sus columnas (tipos de datos y valores no nulos) y un resumen de estadÃ­sticas descriptivas.

```python
import os
import glob
import pandas as pd

directory_path = '/content/drive/MyDrive/FUNPUBLICOS/'

# Use glob to find all files matching the pattern nomina_YYYY-MM.csv
nomina_files = glob.glob(os.path.join(directory_path, 'nomina_*.csv'))

# Initialize an empty list to store individual DataFrames
dfs_list = []

# Iterate through each file_path in the nomina_files list
for file_path in nomina_files:
    try:
        # Read each CSV file into a pandas DataFrame
        df_temp = pd.read_csv(file_path, encoding='latin1', low_memory=False)
        # Append each newly created DataFrame to the dfs_list
        dfs_list.append(df_temp)
    except Exception as e:
        print(f"Error reading {file_path}: {e}")

# Concatenate all DataFrames in the dfs_list into a single DataFrame
combined_df = pd.concat(dfs_list, ignore_index=True)


# Define columns for codes and their descriptions
code_description_pairs = [
    ('codigoNivel', 'descripcionNivel'),
    ('codigoEntidad', 'descripcionEntidad'),
    ('codigoPrograma', 'descripcionPrograma'),
    ('codigoSubprograma', 'descripcionSubprograma')
]

# Create mapping dictionaries/DataFrames for each pair
mapping_tables = {}
for code_col, desc_col in code_description_pairs:
    # Ensure unique mappings. If a code has multiple descriptions, pick the first one.
    mapping_tables[f'{code_col}_to_{desc_col}'] = combined_df[[code_col, desc_col]].drop_duplicates(subset=[code_col]).set_index(code_col)[desc_col].to_dict()
    print(f"Created mapping for {code_col} to {desc_col}. Sample: {list(mapping_tables[f'{code_col}_to_{desc_col}'].items())[:3]}")

# List of columns to keep in the final 'compiled_df'
# Keeping all 'codigo' columns, year/month, amount, and selected abbreviated descriptions
columns_to_keep = [
    'anio',
    'mes',
    'codigoNivel',
    'codigoEntidad',
    'codigoPrograma',
    'codigoSubprograma',
    'montoDevengado',
    'mesCorte',
    'anioCorte',
    'fechaCorte',
    'nivelAbr',
    'entidadAbr',
    'programaAbr',
    'subprogramaAbr',
    'proyectoAbr',
    'unidadAbr'
]

# Create the compiled DataFrame
compiled_df = combined_df[columns_to_keep].copy()


print("\n--- Head of compiled_df ---")
display(compiled_df.head())

print("\n--- Info of compiled_df ---")
compiled_df.info()

print("\n--- Descriptive Statistics of compiled_df ---")
display(compiled_df.describe(include='all'))
```

## Load and Concatenate Data

### Subtask:
Iterate through the discovered file paths. For each file, load it into a pandas DataFrame using `encoding='latin1'` and collect all DataFrames in a list. After processing all files, concatenate them into a single DataFrame. I will also specify `low_memory=False` in `read_csv` to handle potential mixed data types efficiently.

To load and concatenate all discovered `nomina` CSV files into a single DataFrame, I will iterate through the `nomina_files` list, read each file into a DataFrame with the specified encoding and `low_memory` option, append it to a list, and then concatenate all collected DataFrames.



```python
import pandas as pd

# Initialize an empty list to store individual DataFrames
dfs_list = []

# Iterate through each file_path in the nomina_files list
for file_path in nomina_files:
    try:
        # Read each CSV file into a pandas DataFrame
        # Ensure to pass encoding='latin1' and low_memory=False as arguments
        df_temp = pd.read_csv(file_path, encoding='latin1', low_memory=False)
        # Append each newly created DataFrame to the dfs_list
        dfs_list.append(df_temp)
    except Exception as e:
        print(f"Error reading {file_path}: {e}")

# Concatenate all DataFrames in the dfs_list into a single DataFrame
combined_df = pd.concat(dfs_list, ignore_index=True)

# Display the head of the combined DataFrame to verify
print("Combined DataFrame Head:")
display(combined_df.head())

# Display the shape of the combined DataFrame
print(f"\nCombined DataFrame Shape: {combined_df.shape}")
```
